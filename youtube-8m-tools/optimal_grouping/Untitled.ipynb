# This is a notebook that display optimal grouping

```python
labels_list = map(lambda line: map(int, line.split(",")[-1].strip().split()), open("train_labels.csv").readlines())
```

```python
num_crosses = [0] * 4716
loop = 0
for labels in labels_list:
    labels.sort()
    for i in xrange(len(labels)-1):
        l1 = labels[i]
        l2 = labels[i+1]
        weight = (i+1) * (len(labels)-1-i)
        for j in xrange(l1, l2):
            num_crosses[j] += 1
    loop += 1
    if loop % 10000 == 0:
        print loop, "/", len(labels_list)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(len(num_crosses))[:200], num_crosses[:200])
plt.show()
```
